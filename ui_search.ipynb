```python
import tkinter as tk    

root = tk.Tk()    
root.title('5051 project2')    
root.geometry('600x600')


def check1():
    if v_type.get() == True:     
        e_user.configure(state = tk.NORMAL)
    else:
        e_user.configure(state = tk.DISABLED)
        
def check2():
    if trip.get() == True:     
        e_user1.configure(state = tk.NORMAL)
    else:
        e_user1.configure(state = tk.DISABLED)

def check3():
    if gantido.get() == True:     
        e_user2.configure(state = tk.NORMAL)
    else:
        e_user2.configure(state = tk.DISABLED)

        
        
## Search part       
w = tk.Label(root, text="Search", font=("华文行楷", 20), fg="blue")
w.pack()
 

# 1.vehicle type
v_type = tk.BooleanVar()
c1 = tk.Checkbutton(root, text='Vehicle Type', variable=v_type,command=check1)
c1.pack()
# input of the type
e_user =tk.Entry(root, state=tk.DISABLED)
e_user.pack()


# 2.trip length
trip= tk.BooleanVar()
c2 = tk.Checkbutton(root, text='Trip Length', variable=trip,command=check2)
c2.pack()
# input of the trip length
e_user1 =tk.Entry(root, state=tk.DISABLED)
e_user1.pack()


# 2.gantryid_o
gantido= tk.BooleanVar()
c3 = tk.Checkbutton(root, text='GantryID O', variable=gantido,command=check3)
c3.pack()
# input of the id
e_user2 =tk.Entry(root, state=tk.DISABLED)
e_user2.pack()



search_btn = tk.Button(root,text = 'Search',bg = 'red', command = SearchCall)
search_btn.pack()


## Sort part
w1 = tk.Label(root, text="Sort", font=("华文行楷", 20), fg="blue")
w1.pack()
 

dtime_sort = tk.Button(root,text = 'Departure Time',bg = 'red', command = dtime)
dtime_sort.pack()

atime_sort = tk.Button(root,text = 'Arrive Time',bg = 'red', command = atime)
atime_sort.pack()

length_sort = tk.Button(root,text = 'Trip Length',bg = 'red', command = triplen)
length_sort.pack()

ganto_sort = tk.Button(root,text = 'GantryID O',bg = 'red', command = gantryo)
ganto_sort.pack()

gantd_sort = tk.Button(root,text = 'GantryID D',bg = 'red', command = gantryd)
gantd_sort.pack()


root.mainloop()
```
